```python
import matplotlib.pyplot as plt

# Data for the line chart
prompts = ["prompt1", "prompt2", "prompt3"]
student_asks_for_answer = [-0.311, -0.303, -0.265]
rubber_duck_debugging = [0.089, 0.142, 0.108]

# Creating the line chart
plt.figure(figsize=(8, 6))
plt.plot(prompts, student_asks_for_answer, marker='o', label='Student Asks for Answer')
plt.plot(prompts, rubber_duck_debugging, marker='o', label='Rubber Duck Debugging')

# Adding labels, title, and legend
plt.title('Line Chart Representing Values by Prompts')
plt.legend()
plt.grid(True)
plt.ylim(-1, 1)
# Display the graph
plt.tight_layout()
plt.show()
```
